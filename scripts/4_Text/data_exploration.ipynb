```python
# Exploration 17: Add prefixes to column names and prepare for concatenation
print("=" * 60)
print("ADDING PREFIXES TO COLUMN NAMES")
print("=" * 60)

# Create copies to avoid modifying original dataframes
df_3108_prefixed = df_3108.copy()
df_3170_prefixed = df_3170.copy()
df_3370_prefixed = df_3370.copy()

# Add prefix to all columns except patient_id
prefix_3108 = '3108_'
prefix_3170 = '3170_'
prefix_3370 = '3370_'

# Rename columns (keep patient_id as is, prefix all others)
df_3108_prefixed.columns = [col if col == 'patient_id' else f'{prefix_3108}{col}' 
                             for col in df_3108_prefixed.columns]
df_3170_prefixed.columns = [col if col == 'patient_id' else f'{prefix_3170}{col}' 
                             for col in df_3170_prefixed.columns]
df_3370_prefixed.columns = [col if col == 'patient_id' else f'{prefix_3370}{col}' 
                             for col in df_3370_prefixed.columns]

print("\nDataset 3108 columns (first 10):")
print(list(df_3108_prefixed.columns[:10]))

print("\nDataset 3170 columns (first 10):")
print(list(df_3170_prefixed.columns[:10]))

print("\nDataset 3370 columns (first 10):")
print(list(df_3370_prefixed.columns[:10]))

print(f"\nTotal columns in 3108: {len(df_3108_prefixed.columns)}")
print(f"Total columns in 3170: {len(df_3170_prefixed.columns)}")
print(f"Total columns in 3370: {len(df_3370_prefixed.columns)}")
```

```python
# Exploration 18: Concatenate datasets horizontally on patient_id
print("=" * 60)
print("HORIZONTAL CONCATENATION ON PATIENT_ID")
print("=" * 60)

# Merge datasets horizontally using patient_id as the key
# Using outer join to keep all patients from all datasets
df_merged = df_3108_prefixed.merge(
    df_3170_prefixed, 
    on='patient_id', 
    how='outer',
    suffixes=('', '_3170')  # This won't be used since we already prefixed, but good practice
).merge(
    df_3370_prefixed,
    on='patient_id',
    how='outer',
    suffixes=('', '_3370')
)

print(f"\nMerged dataset shape: {df_merged.shape}")
print(f"  Rows: {df_merged.shape[0]}")
print(f"  Columns: {df_merged.shape[1]}")

print(f"\nExpected columns: {len(df_3108_prefixed.columns) + len(df_3170_prefixed.columns) + len(df_3370_prefixed.columns) - 2}")
print(f"  (Subtracting 2 because patient_id appears in all three datasets)")

print("\nFirst few rows of merged dataset:")
print(df_merged.head())

print("\nColumn names in merged dataset:")
print(list(df_merged.columns))
```
